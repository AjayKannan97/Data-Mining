```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
```

```python
cgmdata = pd.read_csv('Project 1 Student Files/CGMData.csv',usecols=['Date','Time','Sensor Glucose (mg/dL)'])
insdata = pd.read_csv('Project 1 Student Files/InsulinData.csv',usecols=['Date','Time','Alarm'])
cgmdata
```

```python
insdata#.columns
```

```python
cgmdata['datetime']=pd.to_datetime(cgmdata['Date'] + ' ' + cgmdata['Time'])
insdata['datetime']=pd.to_datetime(insdata['Date'] + ' ' + insdata['Time'])
cgmdata
```

```python
removeDate=cgmdata[cgmdata['Sensor Glucose (mg/dL)'].isna()]['Date'].unique()
cgmdata=cgmdata.set_index('Date').drop(index=removeDate).reset_index()
cgmdata
```

```python
startAuto=insdata.sort_values(by='datetime',ascending=True).loc[insdata['Alarm']=='AUTO MODE ACTIVE PLGM OFF'].iloc[0]['datetime']
startAuto
```

```python
auto = cgmdata.sort_values(by='datetime',ascending=True).loc[cgmdata['datetime']>=startAuto]
auto
```

```python
manual = cgmdata.sort_values(by='datetime',ascending=True).loc[cgmdata['datetime']<startAuto]
manual
```

```python
auto=auto.set_index('datetime')
l1=auto.groupby('Date')['Sensor Glucose (mg/dL)'].count().where(lambda x:x>0.8*288).dropna().index.tolist()
auto=auto.loc[auto['Date'].isin(l1)]

manual=manual.set_index('datetime')
l2=manual.groupby('Date')['Sensor Glucose (mg/dL)'].count().where(lambda x:x>0.8*288).dropna().index.tolist()
manual=manual.loc[manual['Date'].isin(l2)]
```

```python
#Auto mode
hyperglycemia_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>180].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hyperglycemia_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>180].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hyperglycemia_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>180].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)

hyperglycemia_critical_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>250].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hyperglycemia_critical_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>250].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hyperglycemia_critical_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']>250].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)

range_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=180)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
range_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=180)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
range_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=180)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)

range_sec_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=150)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
range_sec_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=150)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
range_sec_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[(auto['Sensor Glucose (mg/dL)']>=70) & (auto['Sensor Glucose (mg/dL)']<=150)].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)

hypoglycemia_level1_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<70].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hypoglycemia_level1_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<70].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hypoglycemia_level1_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<70].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)

hypoglycemia_level2_wholeday_auto=(auto.between_time('0:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<54].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hypoglycemia_level2_daytime_auto=(auto.between_time('6:00:00','23:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<54].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
hypoglycemia_level2_overnight_auto=(auto.between_time('0:00:00','05:59:59')[['Date','Time','Sensor Glucose (mg/dL)']].loc[auto['Sensor Glucose (mg/dL)']<54].groupby('Date')['Sensor Glucose (mg/dL)'].count()/288*100)
```
